```python
#1

import simpleguitk as simplegui

#event handlers
def print_hello():
    print('Hello')

def print_goodbye():
    print('Goodbye')

#frame
f = simplegui.create_frame("Greetings",300,300)

#register
f.add_button('Hello', print_hello, 100)
f.add_button('Goodbye', print_goodbye, 100)

#start frame
f.start()
```

```python
#2

import simpleguitk as simplegui

color = 'purple'

def print_color():
    print(color)

def set_red():
    global color
    color = 'red'
    print(color)

def set_blue():
    global color
    color = 'blue'
    print(color)

f = simplegui.create_frame('color', 250, 250)

f.add_button('Set red', set_red, 130)
f.add_button('Set blue', set_blue, 130)

f.start()
```

```python
#3

import simpleguitk as simplegui

count = 0

def reset():
    global count
    count = 0
    print(count)

def increment():
    global count
    count += 1
    print(count)

def decrement():
    global count
    count -= 1
    print(count)

f = simplegui.create_frame('+-', 234, 234)

f.add_button('Increment', increment, 123)
f.add_button('Decrement', decrement, 123)
f.add_button('Reset', reset, 123)

f.start()
```

```python
#4

import simpleguitk as simplegui

text = 'ssss'

def echo(s):
    global text
    text = s
    print(text)

f = simplegui.create_frame('Echo', 234, 234)

f.add_input('Input', echo, 200)

f.start()
```

```python
#5

# Rework of Pig Latin, interactive

import simpleguitk as simplegui

word = 'a'

def pig_latin(userinput):
    global word
    first_letter = userinput[0]
    rest_of_word = userinput[1 : ]
    #if first_letter == 'a' or 'e' or 'i' or 'o' or 'u':
    if first_letter == 'a' or first_letter == 'e' or first_letter == 'i' or first_letter == 'o' or first_letter == 'u':
        word = userinput + 'way'
        print(word)
    else:
        word = rest_of_word + first_letter + 'ay'
        print(word)

f = simplegui.create_frame('Pig Latin', 234, 234)

f.add_input('Word', pig_latin, 200)

f.start()
```
